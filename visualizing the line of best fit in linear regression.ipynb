```python
# Learning the slope and intercept by minimizing the error

import numpy as np
import matplotlib.pyplot as plt

# Generating some example data
N = 100  # Number of data points
x = np.random.rand(N, 1) * 10  # Random data for x
y = 3.5 * x + np.random.randn(N, 1) * 2 + 1.5  # Generating y with some noise

m1 = np.random.randn(1,1)
c1 = np.random.randn(1,1)
learningrate = 0.001
numiter = 50

for i in range(numiter):
    # Calculating gradients
    dm = (2.0/N) * -np.sum(np.multiply(x, (y - (m1 * x + c1))))
    dc = (2.0/N) * -np.sum(y - (m1 * x + c1))
    
    # Updating parameters
    m1 = m1 - learningrate * dm
    c1 = c1 - learningrate * dc  # Corrected to use dc for updating c1

    # Printing the number of iterations
    print("no of iterations:", i)
    
    # Plotting data and fit
    plt.plot(x, y, 'go')  # Green dots for data
    plt.plot(x, m1*x + c1, 'r-')  # Red line for the fit

    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Fit at iteration {}'.format(i))
    plt.show()  # Show plot for each iteration

# The final plot
plt.plot(x, y, 'go')
plt.plot(x, m1*x + c1, 'r-', label=f'y = {m1[0][0]:.2f}x + {c1[0][0]:.2f}')
plt.xlabel('x')
plt.ylabel('Data')
plt.legend()
plt.show()
```
